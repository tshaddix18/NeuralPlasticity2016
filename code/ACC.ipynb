```python
def calculate_lowest_difference(bits_range, div_range, subtract_value):
    results = {}
    
    for bits in bits_range:
        max_value = 2**bits - 1
        lowest_result = None
        lowest_diff = float('inf')  # Start with an infinitely large difference
        for div in div_range:
            division_result = (max_value / div) - subtract_value
            # Check if this result is closer to 0 (smallest absolute difference)
            if abs(division_result) < lowest_diff:
                lowest_diff = abs(division_result)
                lowest_result = (div, division_result)
        results[bits] = lowest_result
    
    return results

# Define the range of bits and the range of divisors
bits_range = range(1, 1021)  # Bits 1 to 40
div_range = range(1, 1021)   # Dividing by 1 to 40
subtract_value = 1_200_000

# Calculate the results
lowest_results = calculate_lowest_difference(bits_range, div_range, subtract_value)

# Display the lowest difference for each bit
for bits, result in lowest_results.items():
    divisor, value = result
    print(f"Bits: {bits} | Divided by {divisor}: {value:.2f}")

```
